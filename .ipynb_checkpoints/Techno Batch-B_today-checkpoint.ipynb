```python
## SVM Example and coding


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=pd.read_csv('SVM_IRIS.csv')
```

```python
iris.head()
```

```python
iris.info()
```

```python
plt.scatter(iris['sepal_length'],iris['sepal_width'])
```

```python
plt.scatter(iris['petal_length'],iris['petal_width'])
```

```python
X=iris[['sepal_length','sepal_width','petal_length','petal_width']].values
```

```python
X
```

```python
y=iris[['species']].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.03,random_state=0)
```

```python
from sklearn.svm import SVC
```

```python
sm=SVC()
```

```python
sm.fit(X_train,y_train)
```

```python
y_test
```

```python
y_pred=sm.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# unsupervised ML
# k-mean clustering
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('k-mean_Iris_today.csv')
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,[1,2,3,4]]
```

```python
x.head(3)
```

```python
x=x.values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,16):
    kmean=KMeans(n_clusters=i)
    kmean.fit(x)
    print(kmean.inertia_)
    wcss.append(kmean.inertia_)
    
```

```python
plt.plot(range(1,16),wcss)
plt.xlabel('No of cluster')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
```

```python
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 40, c = 'red', label = 'Iris-A')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 70, c = 'blue', label = 'Iris-B')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 90, c = 'green', label = 'Iris-C')


#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend(loc='upper right')
plt.show()
```
